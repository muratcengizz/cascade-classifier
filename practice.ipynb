```python
import cv2
```

```python
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

def detect_faces(image):
    gray = cv2.cvtColor(src=image, code=cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(image=gray,scaleFactor=1.3, minNeighbors=5)
    
    for (x, y, width, height) in faces:
        cv2.rectangle(img=image,
                      pt1=(x, y),
                      pt2=(x+width, y+height),
                      color=(0, 0, 255),
                      thickness=4)
    return image
    
image = cv2.imread(filename="images/einstein.jpg")
output = detect_faces(image=image)
cv2.imshow(winname="image", mat=image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")

def detect_eyes(image):
    gray = cv2.cvtColor(src=image, code=cv2.COLOR_BGR2GRAY)
    eyes = eye_cascade.detectMultiScale(image=gray, scaleFactor=1.3, minNeighbors=5)
    
    for (x, y, width, height) in eyes:
        cv2.rectangle(img=image,
                      pt1=(x, y),
                      pt2=(x+width, y+height),
                      color=(0, 0, 255),
                      thickness=5)
    return image
    
video = cv2.VideoCapture(0)

while True:
    retval, img = video.read()
    if not retval: break
    img = cv2.flip(src=img, flipCode=1)
    output = detect_eyes(image=img)
    
    
    cv2.imshow(winname="Video", mat=output)
    if cv2.waitKey(1) == ord("q"): break
    
video.release()
cv2.destroyAllWindows()
```
